```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Culture_policy_round2b.csv")
data.head()
data.info()
```

```python
import statsmodels.formula.api as smf

#dropping outliears.
data=data[data["Alpha 0"]>0.01]

#for Appendix, by dropping the next line, one can repeat the analysis for all coutnries.
data=data[data["Too Much Excess Death"]<1]
```

```python
#Regressions Alpha against cultural measures
reg1='np.log(data["Alpha 0"])~data["individualism"]+data["uncertainty avoidance"]+data["power distance"]+data["masculinity"]+data["Long-term orientation"]+data["Indulgence"]'
reg1output=smf.ols(reg1,data).fit()
print(reg1output.summary())
```

```python
##Regressions Alpha against individualism and uncertainity avoidance 
reg1='np.log(data["Alpha 0"])~data["individualism"]+data["uncertainty avoidance"]'
reg1output=smf.ols(reg1,data).fit()
print(reg1output.summary())
```

```python
## Regress death against all cultural measures with and without Alpha
reg1='np.log(data["next_dpm_180"])~np.log(data["Alpha 0"])+data["individualism"]+data["uncertainty avoidance"]+data["power distance"]+data["masculinity"]+data["Long-term orientation"]+data["Indulgence"]'
reg1output=smf.ols(reg1,data).fit()
print(reg1output.summary())

reg1='np.log(data["next_dpm_180"])~data["individualism"]+data["uncertainty avoidance"]+data["power distance"]+data["masculinity"]+data["Long-term orientation"]+data["Indulgence"]'
reg1output=smf.ols(reg1,data).fit()
print(reg1output.summary())

```

```python
#Limiting the dataset to coutnries that have cultural measures.
data=data[data["individualism"]>0]
data=data[data["uncertainty avoidance"]>0]
data=data[data["Indulgence"]>0]

## Regress death against Alpha as a base for comparison
reg1='np.log(data["next_dpm_180"])~np.log(data["Alpha 0"])'
reg1output=smf.ols(reg1,data).fit()
print(reg1output.summary())
```
